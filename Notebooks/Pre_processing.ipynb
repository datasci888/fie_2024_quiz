```python
df_csv.rename(columns={
    'question': 'question',
    'distractor3': 'A',
    'distractor1': 'B',
    'distractor2': 'C',
    'correct_answer': 'D',
    'support': 'context'
}, inplace=True)
```

```python
df_csv['Answer'] = df_csv['D']
df_csv
```

```python
df_csv = df_csv[['question', 'A', 'B', 'C', 'D', 'Answer', 'context']]
```

```python
df_csv.dropna(inplace=True)
df_csv.shape
```

```python
txt_file_path = '/content/drive/MyDrive/Quiz_Generation/SciQdataset/cleaned_data.txt'
with open(txt_file_path, 'w') as file:
    for index, row in df_csv.iterrows():
        file.write(f"Question: {row['question']}\n")
        file.write(f"A: {row['A']}\n")
        file.write(f"B: {row['B']}\n")
        file.write(f"C: {row['C']}\n")
        file.write(f"D: {row['D']}\n")
        file.write(f"Answer: {row['Answer']}\n")
        file.write(f"Context: {row['context']}\n")
        file.write("\n")  # Add a blank line between entries
```
